```python
import cv2
from PIL import Image
import numpy as np
from sklearn import decomposition
import os
```

```python
i=os.getcwd()+"\\Images\\"
data=os.listdir(i)
```

```python
count=0
for i in range(0,len(data),2):
    path1=os.getcwd()+"\\Images\\"+data[i].zfill(3)
    path2=os.getcwd()+"\\Images\\"+data[i+1].zfill(3)
    
    
    pil_image1 = Image.open(path1).convert('RGB') 
    img1 = np.array(pil_image1) 
    pil_image2 = Image.open(path2).convert('RGB') 
    img2 = np.array(pil_image2) 
    if(img1.shape != img2.shape):
        print('=============================================================================')
        print('Image File 1 = '+data[i])
        print('Image File 2 = '+data[i+1])
        print('=============================================================================')
        print('\n')
        continue
    pca = decomposition.PCA(n_components=200)
    p1 = pca.fit_transform(cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY))
    a1=pca.explained_variance_ratio_
    #a1=a1.reshape(a1.shape[0],1)
    print('Image 1 = '+data[i])
    print('Image 2 = '+data[i+1])
    count+=1
    print('Number of Successful passes = '+str(count))
    p2 = pca.fit_transform(cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY))
    a2=pca.explained_variance_ratio_
    #a2=a2.reshape(a2.shape[0],1)
    print('Coeff 1 :'+str(max(a1)))
    print('Coeff 2 :'+str(max(a2)))
    result = cv2.addWeighted(img1,max(a1),img2,max(a2),1)

    #cv2.imshow('IMG1',img1)
    #cv2.imshow('IMG2',img2)
    #cv2.imshow('Result for '+str(data[i]),result)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()

    
    
    outpath=os.getcwd()+"\\Output\\"
    cv2.imwrite(outpath+str(data[i]).zfill(3)+".jpg", result)
    
    #####################################
    result1 = cv2.addWeighted(img1,0.5,img2,0.5,0)
    cv2.imwrite(outpath+str(data[i]).zfill(3)+"_Auto_.jpg", result1)
    
    #cv2.imshow('IMG1',img1)
    #cv2.imshow('IMG2',img2)
    #cv2.imshow('Result',result)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    print('----------------------------------------------------------------------------------------------------')
```
